```python
#creating a new dataframe with columns transposed to rows to find 
top_10_values = foxnews.set_index('Show').stack().reset_index()
top_10_values.columns = ['Show', 'Date', 'Value']
top_10_values = top_10_values.nlargest(10, 'Value')

# Print the top 10 values
print(top_10_values)
```

```python
#export to new csv
top_10_values.to_csv("top10episodes.csv")
```
