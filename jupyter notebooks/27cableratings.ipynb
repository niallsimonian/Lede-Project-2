```python
#imports
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#fetch url
url = "https://www.mediaite.com/daily-ratings/cable-news-ratings-tuesday-june-27-gutfeld-tops-cable-news-in-the-demo/"
base = requests.get(url)
base
```

```python
#convert to html text & test
base_html = base.text 
print(base_html)
```

```python
#make soup object
soup = BeautifulSoup(base_html)
type(soup)
```

```python
#selecting just the first of two tables on the page
table = soup.select("table")[0]
```

```python
#creating header list
first_row = table.select("tbody tr")[0]
header_list = [ cell.text.rstrip("\n") for cell in first_row.select("td") ]
header_list
```

```python
#selecting non header rows of data from table 
rows_ratings = table.select("tr")[1:19]
len(rows_ratings)
print(rows_ratings)
```

```python
rating_entries = []
for row in rows_ratings:
    row_cells = []
    for cell in row.select("td"):
        row_cells.append(cell.text.rstrip("\n"))
    rating_entries.append(row_cells)
rating_entries[:3]
```

```python
#creatingdataframe
cable27_df = pd.DataFrame(rating_entries, columns=header_list)
cable27_df.tail()
```

```python
#dropping everything but fox to create fox column only df
fox27 = cable27_df.drop(columns=['CNN', 'MSNBC','NEWSMAX',"NEWSNATION","ET"])
fox27
```

```python
#splitting the string and numbers in the FNC column, while dropping the :\n . And then dropping the original FNC Column
fox27[['Show', '6/27']] = fox27['FNC'].str.split(':\n', expand=True)
fox27 = fox27.drop(columns=['FNC'])
fox27
```

```python
fox27.to_csv("fox27.csv")
```
